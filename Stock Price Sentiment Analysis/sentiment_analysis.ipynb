```python
import pandas as pd
```

```python
data = pd.read_csv(r"D:\NLP\NLP_FROM_BASICS\Stock Price Sentiment Analysis\Data.csv", encoding="ISO-8859-1")
```

```python
data.head()
```

```python
data.tail(3)
```

```python
train = data[data['Date']<'20150101']
test = data[data['Date']>'20151231']
```

```python
print(train.shape)
print(test.shape)
```

```python
x=train.iloc[:,2:27]
x.replace("[^a-zA-Z]"," ",regex=True,inplace=True)
```

```python
x.head(2)
```

```python
lst = [str(i) for i in range(25)]
x.columns=lst
x.head(2)
```

```python
for col in x.columns:
    x[col]=x[col].str.lower()
x.head(1)
```

```python
print(x.index)
```

```python
#concating headlines
news=[]
for row in range(0,len(x.index)):
    tmp = ' '.join(str(x) for x in x.iloc[row,0:25])
    news.append(tmp)
```

```python
len(news)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(ngram_range=(2,2))
train_vect = vect.fit_transform(news)
```

```python
train_vect.toarray()[1]
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100,criterion='entropy')
model.fit(train_vect,train['Label'])
```

```python
test_trans=[]
for row in range(0,len(test.index)):
    test_trans.append(" ".join(str(i) for i in test.iloc[row,2:27]))
    
test_vect = vect.transform(test_trans)
pred = model.predict(test_vect)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
print(classification_report(test['Label'], pred))
```

```python
confusion_matrix(test.Label,pred)
```

```python
accuracy_score(test.Label,pred)
```

```python
print(f'train accuracy is {confusion_matrix(train.Label,model.predict(train_vect))}')
```

```python
print(f'test accuracy is {confusion_matrix(test.Label,model.predict(test_vect))}')
```
